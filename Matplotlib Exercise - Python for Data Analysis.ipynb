## Subplots Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b =  a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
type(axes)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][0].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
    ax.plot(x,y)
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.9)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2, figsize=(8,5), dpi=200)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(a,b)
axes[1][0].set_ylabel("Y LABEL 1,0",fontsize=10)
axes[1][0].set_xlim(2,6)


axes[1][1].plot(a,b)
axes[1][1].set_title("TITLE 1,1",fontsize=10)

fig.suptitle("Figure Level", fontsize=16)
fig.tight_layout()

fig.savefig("new_subplots.png", bbox_inches="tight")
#fig.subplots_adjust(wspace=0.35,hspace=0.2)
```

## Matplotlib Styling 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label="X vs X")

ax.plot(x,x**2, label="X vs X^2")

ax.legend(loc=(1.1,1));
```

## Matplotlib Styling Colors and Styles

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color="#650dbd", label="X vs X") # RGB HEX Code

ax.plot(x,x+1, color="red" , label = " X vs X +1 ") # RGB HEX Code

ax.legend();
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color="#650dbd", lw=10, ls="--") # RGB HEX Code

ax.plot(x,x+1, color="red" , lw=0.5, ls='-.') # RGB HEX Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x, color="#650dbd", lw=10, ls="--") # RGB HEX Code

lines = ax.plot(x,x+1, color="red" , lw=5) # RGB HEX Code

lines[0].set_dashes([5,2,10,2])
```

```python
type(lines)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x, color="#650dbd", lw=10, ls="--") # RGB HEX Code

lines = ax.plot(x,x, color="purple" , lw=2,marker='+',ls='--',ms=10) # RGB HEX Code

lines = ax.plot(x,x+1, color="red" , lw=2,marker='o',ls='--', markersize=10) # RGB HEX Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x, color="#650dbd", lw=10, ls="--") # RGB HEX Code

#lines = ax.plot(x,x, color="purple" , lw=2,marker='+',ls='--',ms=10) # RGB HEX Code

lines = ax.plot(x,x+1, color="red" , lw=2,marker='o',ls='--', markersize=10,
               markerfacecolor="red",markeredgewidth=4,markeredgecolor="orange") # RGB HEX Code
```
